```python
import os
import lightgbm as lgb
import pandas as pd
import numpy as np
from sklearn.model_selection import GroupKFold
from sklearn.metrics import ndcg_score
import matplotlib.pyplot as plt

print("Install Succesfully")
```

```python
import os
import pandas as pd

# ===============================
# Paths & Config
# ===============================
DATASET_PATH = "data/supplier_ranking_v5_large.csv"   # Use the large dataset
MODEL_DIR = "models"
MODEL_PATH = os.path.join(MODEL_DIR, "supplier_ranker_model.txt")
os.makedirs(MODEL_DIR, exist_ok=True)

FEATURES = [
    "price", "lead_time", "on_time_delivery_rate", "quality_acceptance_rate",
    "defect_rate", "past_performance_score", "fulfillment_rate", 
    "responsiveness_score", "capacity_reliability_score", "has_iso_9001",
    "financial_stability_score", "geo_risk_score"
]
LABEL = "relevance"
GROUP_ID = "query_id"

# ===============================
# Load Data with Validation
# ===============================
try:
    df = pd.read_csv(DATASET_PATH)
    print(f"✅ Loaded dataset with {len(df)} rows and {df[GROUP_ID].nunique()} queries")

    # Check required columns
    missing_cols = [col for col in FEATURES + [LABEL, GROUP_ID] if col not in df.columns]
    if missing_cols:
        raise ValueError(f"⚠️ Missing columns in dataset: {missing_cols}")

    # Show basic info
    print(f"Dataset shape: {df.shape}")
    print(f"Unique queries: {df[GROUP_ID].nunique()}")
    print(f"Avg suppliers per query: {df.groupby(GROUP_ID).size().mean():.1f}")

    print("\n--- Relevance Distribution ---")
    print(df[LABEL].value_counts(normalize=True).sort_index())

except FileNotFoundError:
    raise FileNotFoundError(f"❌ Dataset not found at '{DATASET_PATH}'")

X = df[FEATURES]
y = df[LABEL]

print("\n--- Data Ready for Training ---")
print(f"Features shape: {X.shape}")
print(f"Labels shape: {y.shape}")
print(f"Missing values in features: {X.isnull().sum().sum()}")
```

```python
# ===============================
# DATA VALIDATION
# ===============================
X = df[FEATURES]
y = df[LABEL]

print("\n--- Data Validation ---")
print(f"Features shape: {X.shape}")
print(f"Labels shape: {y.shape}")
print(f"Missing values in features: {X.isnull().sum().sum()}")
print(f"Label range: [{y.min()}, {y.max()}]")

# ===============================
# MODEL CONFIG (stronger, tuned for larger dataset)
# ===============================
params = {
    "objective": "lambdarank",
    "metric": "ndcg",
    "boosting_type": "gbdt",
    "n_estimators": 3000,         # more boosting rounds
    "learning_rate": 0.03,        # smaller steps
    "num_leaves": 127,            # allow more splits per tree
    "min_child_samples": 30,      # slightly higher to avoid noise
    "max_depth": -1,
    "colsample_bytree": 0.8,
    "subsample": 0.8,
    "subsample_freq": 1,
    "lambda_l1": 0.1,             # L1 regularization
    "lambda_l2": 0.1,             # L2 regularization
    "seed": 42,
    "n_jobs": -1,
    "verbose": -1,
}

# ===============================
# CROSS VALIDATION TRAINING
# ===============================
gkf = GroupKFold(n_splits=5)
models, fold_ndcg_scores, feature_importance_list = [], [], []

print("\n--- Training with 5-Fold GroupKFold ---")
for fold, (train_idx, val_idx) in enumerate(gkf.split(X, y, groups=df[GROUP_ID])):
    print(f"\n--- Fold {fold+1}/5 ---")

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    train_groups = df.iloc[train_idx].groupby(GROUP_ID).size().to_numpy()
    val_groups = df.iloc[val_idx].groupby(GROUP_ID).size().to_numpy()

    ranker = lgb.LGBMRanker(**params)
    ranker.fit(
        X_train, y_train, group=train_groups,
        eval_set=[(X_val, y_val)], eval_group=[val_groups],
        eval_at=[1, 3, 5, 10],
        callbacks=[lgb.early_stopping(stopping_rounds=100, verbose=True)]
    )
    models.append(ranker)

    # --- Evaluation ---
    y_pred_scores = ranker.predict(X_val)
    validation_df = df.iloc[val_idx].copy()
    validation_df["pred_score"] = y_pred_scores

    query_ndcg_scores = []
    for query_id in validation_df[GROUP_ID].unique():
        query_df = validation_df[validation_df[GROUP_ID] == query_id]
        true_relevance = [query_df[LABEL].tolist()]
        pred_scores = [query_df["pred_score"].tolist()]
        query_ndcg = ndcg_score(true_relevance, pred_scores, k=5)
        query_ndcg_scores.append(query_ndcg)

    avg_ndcg_for_fold = np.mean(query_ndcg_scores)
    fold_ndcg_scores.append(avg_ndcg_for_fold)
    print(f"Fold {fold+1} Avg NDCG@5: {avg_ndcg_for_fold:.4f}")

    # --- Sanity check: top-3 suppliers for one query ---
    sample_query = validation_df[GROUP_ID].iloc[0]
    sample_df = validation_df[validation_df[GROUP_ID] == sample_query].copy()
    sample_df = sample_df.sort_values("pred_score", ascending=False)
    print("\nSample ranking (top 3):")
    print(sample_df[["price", "lead_time", LABEL, "pred_score"]].head(3))

    # --- Feature Importance ---
    feature_importance_list.append(pd.Series(ranker.feature_importances_, index=FEATURES))
```

```python
# ===============================
# CELL 4: MODEL EVALUATION SUMMARY
# ===============================
mean_ndcg, std_ndcg = np.mean(fold_ndcg_scores), np.std(fold_ndcg_scores)
print("\n--- Model Evaluation Summary ---")
print(f"NDCG@5 per fold: {[f'{s:.4f}' for s in fold_ndcg_scores]}")
print(f"Mean NDCG@5: {mean_ndcg:.4f} (+/- {std_ndcg:.4f})")

if mean_ndcg > 0.95:
    print("✅ Excellent model performance")
elif mean_ndcg > 0.85:
    print("✅ Good performance")
elif mean_ndcg > 0.70:
    print("⚠️ Fair performance - consider tuning or better data")
else:
    print("❌ Poor performance - revisit features/labels")

# ===============================
# FEATURE IMPORTANCE
# ===============================
avg_importance = pd.concat(feature_importance_list, axis=1).mean(axis=1).sort_values(ascending=False)
print("\n--- Top 5 Features ---")
for i, (feature, importance) in enumerate(avg_importance.head(5).items()):
    print(f"{i+1}. {feature}: {importance:.1f}")

plt.figure(figsize=(8, 5))
avg_importance.plot(kind="barh")
plt.title("Average Feature Importance (across folds)")
plt.xlabel("Importance Score")
plt.gca().invert_yaxis()
plt.show()
```

```python
import joblib

# Save each fold model separately
for i, model in enumerate(models):
    model_path = os.path.join(MODEL_DIR, f"supplier_ranker_fold{i+1}.pkl")
    joblib.dump(model, model_path)
    print(f"✅ Saved model for Fold {i+1} at: {model_path}")

# Save ensemble (all models together for easy loading later)
ensemble_path = os.path.join(MODEL_DIR, "supplier_ranker_ensemble.pkl")
joblib.dump(models, ensemble_path)
print(f"✅ Saved ensemble of {len(models)} models at: {ensemble_path}")
```

```python
# ===============================
# CELL 6: INFERENCE / PREDICTION
# ===============================
import joblib

# Load saved ensemble
ensemble_path = os.path.join(MODEL_DIR, "supplier_ranker_ensemble.pkl")
models = joblib.load(ensemble_path)
print(f"✅ Loaded ensemble with {len(models)} models")

def rank_suppliers(new_df, features=FEATURES, group_id=GROUP_ID):
    """
    Rank suppliers for new queries using the ensemble.
    - new_df must contain at least FEATURES + [query_id]
    """

    X_new = new_df[features]

    # Predict with each model and average scores
    all_preds = np.zeros((len(new_df), len(models)))
    for i, model in enumerate(models):
        all_preds[:, i] = model.predict(X_new)
    avg_preds = all_preds.mean(axis=1)

    new_df = new_df.copy()
    new_df["pred_score"] = avg_preds

    # Rank within each query
    ranked = []
    for qid, group in new_df.groupby(group_id):
        group = group.sort_values("pred_score", ascending=False).reset_index(drop=True)
        group["rank"] = range(1, len(group) + 1)
        ranked.append(group)
    return pd.concat(ranked)

# ===============================
# TEST ON A SAMPLE QUERY
# ===============================
sample_query_id = df[GROUP_ID].iloc[0]
sample_df = df[df[GROUP_ID] == sample_query_id].copy()

print(f"\n--- Ranking Suppliers for Query {sample_query_id} ---")
ranked_suppliers = rank_suppliers(sample_df)
print(ranked_suppliers[["supplier_id", "price", "lead_time", LABEL, "pred_score", "rank"]])
```
